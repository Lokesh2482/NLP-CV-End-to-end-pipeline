```python
import gc
import torch
import weakref

def cleanup_gpu_memory(obj=None, verbose: bool = False):

    if not torch.cuda.is_available():
        if verbose:
            print("[INFO] CUDA is not available. No GPU cleanup needed.")
        return

    def get_memory_stats():
        allocated = torch.cuda.memory_allocated()
        reserved = torch.cuda.memory_reserved()
        return allocated, reserved

    torch.cuda.synchronize()

    if verbose:
        alloc, reserv = get_memory_stats()
        print(f"[Before] Allocated: {alloc / 1024**2:.2f} MB | Reserved: {reserv / 1024**2:.2f} MB")

    # Ensure we drop all strong references
    if obj is not None:
        ref = weakref.ref(obj)
        del obj
        if ref() is not None and verbose:
            print("[WARNING] Object not fully garbage collected yet.")

    gc.collect()
    torch.cuda.empty_cache()
    torch.cuda.ipc_collect()

    torch.cuda.synchronize()

    if verbose:
        alloc, reserv = get_memory_stats()
        print(f"[After]  Allocated: {alloc / 1024**2:.2f} MB | Reserved: {reserv / 1024**2:.2f} MB")
```

```python
# -*- coding: utf-8 -*-
import json
import os
import csv
from PIL import Image
import numpy as np
import supervision as sv
from rfdetr import RFDETRNano
from tqdm import tqdm
import torch

# ----------------------------
# Paths
# ----------------------------
test_images_path = "/content/test_data"
prompts_json_path = "/content/simplified_test_prompts.json"
output_csv_path = "/content/test_predictions6.csv"
output_vis_path = "/content/test_visualization6"

os.makedirs(output_vis_path, exist_ok=True)

# ----------------------------
# Load test prompts
# ----------------------------
with open(prompts_json_path, "r") as f:
    test_prompts = json.load(f)["images"]

# ----------------------------
# Initialize RF-DETR with trained weights
# ----------------------------
print("[INFO] Loading model...")
model = RFDETRNano(
    num_classes=4,
    pretrain_weights="output/checkpoint_best_total.pth"
)

print("[INFO] Model loaded successfully")

# Use the correct class mapping based on COCO training data
CLASS_NAME_TO_ID = {
    "bulbasaur": 1,
    "charizard": 2,
    "mewtwo": 3,
    "pikachu": 4
}

MODEL_CLASS_ID_TO_NAME = {
    1: "bulbasaur",
    2: "charizard",
    3: "mewtwo",
    4: "pikachu"
}

print(f"[INFO] Final class mapping: {CLASS_NAME_TO_ID}")

# ----------------------------
# Helper functions
# ----------------------------
def compute_center_of_mass(bbox):
    """Compute center point of bounding box"""
    x_min, y_min, x_max, y_max = bbox
    return [(x_min + x_max) / 2, (y_min + y_max) / 2]

def filter_detections_by_class(detections, target_class_id):
    """Filter detections to only include target class"""
    if len(detections.class_id) == 0:
        return []

    # Create mask for target class
    mask = detections.class_id == target_class_id

    if not np.any(mask):
        return []

    # Return filtered bounding boxes
    return detections.xyxy[mask]

# ----------------------------
# Run inference and build CSV + save annotated images
# ----------------------------
csv_rows = []
successful_predictions = 0
class_detection_stats = {name: 0 for name in CLASS_NAME_TO_ID.keys()}

print(f"[INFO] Processing {len(test_prompts)} test images...")

for item in tqdm(test_prompts, desc="Processing test images"):
    image_id = item["id"]
    target_class_name = item["target"][0].lower()

    if target_class_name not in CLASS_NAME_TO_ID:
        print(f"[WARNING] Unknown target class: {target_class_name}")
        csv_rows.append([f"{image_id}.png", json.dumps([])])
        continue

    target_class_id = CLASS_NAME_TO_ID[target_class_name]
    image_filename = f"{image_id}.png"
    image_path = os.path.join(test_images_path, image_filename)

    if not os.path.exists(image_path):
        print(f"[WARNING] Image not found: {image_path}")
        csv_rows.append([image_filename, json.dumps([])])
        continue

    try:
        # Load and process image
        image = Image.open(image_path).convert('RGB')

        # Run inference - try multiple thresholds if needed
        detections = model.predict(image, threshold=0.2)

        # Debug info for first few images
        if len(csv_rows) < 3:
            print(f"[DEBUG] Image {image_id}:")
            print(f"  Target: {target_class_name} (ID: {target_class_id})")
            print(f"  Detected class IDs: {list(detections.class_id)}")
            print(f"  Confidences: {list(detections.confidence)}")

        # Filter by target class
        target_bboxes = filter_detections_by_class(detections, target_class_id)

        if len(target_bboxes) > 0:
            successful_predictions += 1
            class_detection_stats[target_class_name] += 1

            # Ensure target_bboxes is numpy array
            target_bboxes = np.array(target_bboxes, dtype=np.float32)

            # Compute centers for CSV
            targeting_points = [compute_center_of_mass(bbox) for bbox in target_bboxes]
            targeting_points = [[float(x), float(y)] for x, y in targeting_points]
            csv_rows.append([image_filename, json.dumps(targeting_points)])

            # Create visualization
            try:
                filtered_detections = sv.Detections(
                    xyxy=target_bboxes,
                    class_id=np.array([target_class_id] * len(target_bboxes)),
                    confidence=np.ones(len(target_bboxes))
                )

                detections_labels = [f"{target_class_name} {i+1}" for i in range(len(target_bboxes))]

                # Calculate optimal visualization parameters
                text_scale = sv.calculate_optimal_text_scale(resolution_wh=image.size)
                thickness = sv.calculate_optimal_line_thickness(resolution_wh=image.size)

                # Use different colors for different classes
                class_colors = {
                    "pikachu": "#ffff00",    # Yellow
                    "charizard": "#ff4500",  # Orange-red
                    "bulbasaur": "#32cd32",  # Green
                    "mewtwo": "#9932cc"      # Purple
                }
                color_hex = class_colors.get(target_class_name, "#ff0000")
                color = sv.ColorPalette.from_hex([color_hex])

                # Create annotators
                bbox_annotator = sv.BoxAnnotator(color=color, thickness=thickness)
                label_annotator = sv.LabelAnnotator(
                    color=color,
                    text_color=sv.Color.BLACK,
                    text_scale=text_scale
                )

                # Annotate image
                annotated_image = np.array(image.copy())
                annotated_image = bbox_annotator.annotate(annotated_image, filtered_detections)
                annotated_image = label_annotator.annotate(annotated_image, filtered_detections, detections_labels)

                # Save annotated image
                annotated_image_pil = Image.fromarray(annotated_image)
                annotated_image_pil.save(os.path.join(output_vis_path, image_filename))
            except Exception as viz_error:
                print(f"[WARNING] Visualization failed for {image_filename}: {viz_error}")
        else:
            csv_rows.append([image_filename, json.dumps([])])

    except Exception as e:
        print(f"[ERROR] Failed to process {image_filename}: {str(e)}")
        csv_rows.append([image_filename, json.dumps([])])

# ----------------------------
# Save CSV
# ----------------------------
with open(output_csv_path, "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerow(["image_id", "points"])
    writer.writerows(csv_rows)

# ----------------------------
# Print summary statistics
# ----------------------------
print(f"\n[INFO] Processing complete!")
print(f"[INFO] CSV saved at {output_csv_path}")
print(f"[INFO] Annotated images saved at {output_vis_path}")
print(f"[INFO] Images with successful target detections: {successful_predictions}/{len(test_prompts)}")
print(f"[INFO] Detection stats by class: {class_detection_stats}")

# Check if any class has 0 detections
zero_detection_classes = [name for name, count in class_detection_stats.items() if count == 0]
if zero_detection_classes:
    print(f"[WARNING] Classes with no detections: {zero_detection_classes}")
    print("[INFO] Consider lowering the detection threshold or retraining the model")
```
